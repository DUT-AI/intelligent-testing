```python
import json
import time
from googletrans import Translator

# Khởi tạo Translator
translator = Translator()

def translate_text(text, src='zh-cn', dest='vi'):
    if not text or not isinstance(text, str):
        return text
    
    # Giữ nguyên LaTeX và ID hình ảnh
    if "$$" in text or "image_" in text:
        return text
    
    try:
        # Thêm delay nhỏ để tránh bị Google quét (Rate Limit)
        time.sleep(0.5) 
        result = translator.translate(text, src=src, dest=dest)
        return result.text
    except Exception as e:
        print(f"Lỗi khi dịch cụm: {text[:20]}... | Lỗi: {e}")
        return text

def process_json(data):
    """Đệ quy để duyệt toàn bộ cấu trúc JSON"""
    if isinstance(data, dict):
        return {k: process_json(v) for k, v in data.items()}
    elif isinstance(data, list):
        return [process_json(i) for i in data]
    elif isinstance(data, str):
        return translate_text(data)
    else:
        return data

# --- THỰC THI ---
input_filename = r'E:\Semester 8 Document\IntelligentTesting\XES3G5M\XES3G5M\metadata\question_readable_full.json' # Đảm bảo file này nằm cùng thư mục với notebook
output_filename = r'E:\Semester 8 Document\IntelligentTesting\XES3G5M\XES3G5M\metadata\questions_translated.json'

try:
    with open(input_filename, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    print("Bắt đầu dịch... Tiến trình này có thể mất vài phút tùy độ lớn của file.")
    
    translated_data = process_json(data)
    
    with open(output_filename, 'w', encoding='utf-8') as f:
        json.dump(translated_data, f, ensure_ascii=False, indent=4)
        
    print(f"Thành công! File đã lưu tại: {output_filename}")
except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy file '{input_filename}'. Hãy upload file lên thư mục hiện tại.")
except Exception as e:
    print(f"Đã xảy ra lỗi hệ thống: {e}")
```
